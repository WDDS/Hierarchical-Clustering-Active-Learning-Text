```python
model = KeyedVectors.load_word2vec_format("/home/adrian/PhD/Data/Word2Vec/BioASQvectors2018/pubmed2018_w2v_200D/pubmed2018_w2v_200D.bin", binary=True)
index2word_set = set(model.wv.index2word)

def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

def avg_feature_vector(sentence, model=model, num_features=200, index2word_set=index2word_set):

    feature_vec = np.zeros((num_features, ), dtype='float32')
    n_words = 0
    for word in sentence:
        if word in index2word_set:
            n_words += 1
            feature_vec = np.add(feature_vec, model[word])
        else:
            if hasNumbers(word):
                print("word not in vocabulary: {}".format(word))
    if (n_words > 0):
        feature_vec = np.divide(feature_vec, n_words)
    return feature_vec


annotationsData["vec"] = annotationsData.tokens.map(lambda token: avg_feature_vector(token))
annotationsData.head()
```

```python
training = np.stack(annotationsData["vec"].values, axis = 0)
modelSVM = sklearn.svm.LinearSVC(C=0.01, max_iter=100, fit_intercept=True, tol=1e-6)
modelSVM.fit(X = training
         , y = annotationsData["labels"].map(lambda label: 1 if label == True else 0))
```

```python
import pyspark
from pyspark.sql.functions import udf, col

spark = pyspark.sql.SparkSession.builder.getOrCreate()
sentences = spark.read.load("/home/adrian/workspace/FBE output_without_root_class_diabetesMellitus/maxTopwords_6_Nall/phrases/")
print(len(sentences.columns))
print(sentences.count())

phrases = sentences.select("id", "tokens", "index")
phrases.show(2)
```

```python
index_1_udf = udf(lambda ind: list(ind[1].keys()))
tokensInScope_udf = udf(lambda tokens, index: [tokens[i] for i in sorted(list(index[1].keys()))] )
join_udf = udf(lambda x: ";".join(x))
phrasesNew = phrases.select("id"
              , "tokens"
              , tokensInScope_udf("tokens", "index").alias("tokensInScope")
              , index_1_udf("index").alias("tokens_in_key_1")) \
        .withColumn("tokens", join_udf(col("tokens"))) \
        .withColumn("tokensInScope", join_udf(col("tokensInScope")))

#    sentences_classes_udf = udf(lambda x: ";".join([str(v) for v in x.keys()]))
phrasesNew.show(2)

phrases_pd = phrasesNew.toPandas()
```

```python
def ff(row):
    tokens = set(row["tokens"].split(";"))
    tokensInScope = set(row["tokensInScope"].split(";"))
    return (list(tokens-tokensInScope))
    
outScopeWords = phrases_pd.apply(ff, axis=1).explode().unique()
print(len(outScopeWords))
```
